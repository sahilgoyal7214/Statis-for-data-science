```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as pit
```

```python
song = pd.read_csv('/content/song_data.csv')
song
```

```python
song.index=song.iloc[:,0]
```

```python
song
```

```python
song=song.drop(['song_name'],axis=1)
```

```python
song.head()
```

```python
pit.figure(figsize=(15,15))
sb.heatmap(song.corr(),annot=True,linewidth=3)
```

```python
input_x = song['danceability']
output_y = song['song_popularity']
np.array(input_x)
np.array(output_y)
```

```python
def weight(input,output):
  mean_x=np.mean(input)
  mean_y=np.mean(output)
  theta_1=(np.sum((input-mean_x)*(output-mean_y)))/(np.sum((input-mean_x)**2))
  
  theta_0=mean_y-theta_1*mean_x
  return[theta_0,theta_1]
```

```python
t_0,t_1=weight(input_x,output_y)
print(t_0,t_1)
```

```python
y_bar=t_0 + t_1*input_x
y_bar
```

```python
rss=np.sum((output_y-y_bar)**2)
rss
```

```python
mse=np.mean((output_y-y_bar)**2)
mse
```

```python
rmse=mse**(1/2)
rmse
```

```python
mae=np.mean(abs(y_bar-output_y))
mae
```

```python
mape=np.mean(abs(y_bar-output_y)/(output_y+0.001))*100
mape
```

```python
pit.scatter(output_y,y_bar)
pit.show()
```
